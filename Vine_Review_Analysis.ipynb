```python
import pandas as pd
```

```python
vine_table = pd.read_csv('vine_table.csv')
```

```python
vine_table.head()
```

```python
# Table with total vote count greater than 20
vc20_df = vine_table[vine_table.total_votes >= 20]
vc20_df.head()
```

```python
# Create column to calculate percentage of helpful votes with respect to total votes
vc20_df['pct_helpful_votes'] = vc20_df['helpful_votes'] / vc20_df['total_votes']
fifty_pct_df = vc20_df[vc20_df.pct_helpful_votes >= 0.5]
fifty_pct_df.head()
```

```python
# Create table where vine is yes
vine_y_df = fifty_pct_df[fifty_pct_df.vine == 'Y']
vine_y_df.head()
```

```python
# Create table where vine is yes
vine_n_df = fifty_pct_df[fifty_pct_df.vine == 'N']
vine_n_df.head()
```

```python
#Summarize paid reviews
paid_summary = vine_y_df.groupby('star_rating').agg({'review_id': 'count'})
paid_summary.columns = ['count_of_reviews']
paid_summary['pct_of_total'] = paid_summary.count_of_reviews / sum(paid_summary.count_of_reviews)
paid_summary
```

```python
print('Number of paid reviews: ', sum(paid_summary.count_of_reviews))
```

```python
#Summarize unpaid reviews
unpaid_summary = vine_n_df.groupby('star_rating').agg({'review_id': 'count'})
unpaid_summary.columns = ['count_of_reviews']
unpaid_summary['pct_of_total'] = unpaid_summary.count_of_reviews / sum(unpaid_summary.count_of_reviews)
unpaid_summary
```

```python
print('Number of unpaid reviews: ', sum(unpaid_summary.count_of_reviews))
```
